```python
# Importing the required libraries:

import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display
```

# Example spectrograms from the dataset used:

```python
audio, sr = librosa.load("crash_1.wav", sr=None)

plt.figure(figsize=(15,5))

spec = librosa.feature.melspectrogram(y=audio, sr=sr)

librosa.display.specshow(librosa.power_to_db(spec, ref=np.max), x_axis='time', y_axis='mel')

plt.colorbar(format='%+2.0f dB')
plt.title("Crash Audio Spectrogram (Example 1)")

plt.tight_layout()
plt.show()
```

```python
audio, sr = librosa.load("crash_14.wav", sr=None)

plt.figure(figsize=(15,5))

spec = librosa.feature.melspectrogram(y=audio, sr=sr)

librosa.display.specshow(librosa.power_to_db(spec, ref=np.max), x_axis='time', y_axis='mel')

plt.colorbar(format='%+2.0f dB')
plt.title("Crash Audio Spectrogram (Example 2)")

plt.tight_layout()
plt.show()
```

# Spectrogram Analysis of a Crash Audio Sample

The spectrogram provides a visual representation of how the frequency content of an audio signal changes over time. In this crash audio sample, a noticeable high-energy spike can be observed at the moment of impact.

This sudden spike appears because a vehicle collision generates a large amount of acoustic energy within a very short duration. The impact produces sharp transient sounds that spread across a wide range of frequencies, creating bright vertical patterns in the spectrogram.

Compared to normal environmental sounds, crash events tend to contain abrupt changes in amplitude and frequency distribution. These distinctive characteristics make spectrograms useful for understanding why machine learning models can differentiate crash sounds from ordinary background noise.

The spectrogram shown here is generated from an actual crash audio sample used during the development and testing of the crash detection system.
